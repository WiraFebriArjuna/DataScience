```
import pandas as pd, requests
```

##CURRENCY FLUCTUATION

```
url = 'https://api.exchangerate.host/fluctuation?'
parameters = {
        'start_date':'2021-01-22',  
        'end_date' : '2021-02-22',
        'base': 'USD'
}
response = requests.get(url, params=parameters)
response.status_code
```

```
fluctuation = response.json()
fluctuation
```

```
start_date = fluctuation['start_date']
end_date = fluctuation['end_date']
changes = fluctuation['rates']
temp = []
for currency in changes:
    start_rate  = changes[currency]['start_rate']
    end_rate    = changes[currency]['end_rate']
    change      = changes[currency]['change']
    change_pct  = changes[currency]['change_pct']
    row = [start_date, end_date, currency, start_rate, end_rate, change, change_pct]
    temp.append(row)

df_fluctuation = pd.DataFrame(temp,
                              columns = ['start_date', 'end_date', 'currency', 'start_rate', 'end_rate', 'change', 'change_pct'])
df_fluctuation
```
